### Creating a empty Series

```python
import numpy as np
import pandas as pd 
ser=pd.Series()
print(ser)
```

##### Creating a series from array:

```python
import numpy as np
import pandas as pd 
data=np.array(['N','A','V','E','E','N','A'])
series=pd.Series(data)
print(series)
```

##### Creating a series from array with an index

```python
data=np.array(['N','A','V','E','E','N','A'])
ser=pd.Series(data,index=[10,11,12,13,14,15,16])
print(ser)
```

##### Creating a Series from list 

```python
import numpy as np
import pandas as pd 
list=[11,2,3,4,5,6,7,8,9,11]
ser=pd.Series(list)
print("Series",ser)
print("list:",list)
```

##### Creating a series from Dictionary

```python
import numpy as np
import pandas as pd 
dict={'Name':'Naveenasri','dept':'CSE','clg':'KPR'}
series=pd.Series(dict)
print(series)
print("dictionary:",dict)
```

##### Range function

##### Creating a series using Mathematical expression

##### Accessing a elements in series

```python
import pandas as pd
import numpy as np
data =np.array([100,22,44,55,66])
ser=pd.Series(data)
print(ser[:4])
```

```python
data =np.array([100,22,44,55,66,88,99,00,33,4,5,6,7,8,9])
ser=pd.Series(data)
print(ser[1:4])
print(ser[1:])
print(ser[-5:])
print(ser[5])
print(ser[5:])
```

##### Accessing Element using lable(index)

```python
import numpy as np
import pandas as pd 
data=np.array(['naveena','mohana','kani','kpr','cse','k','t','m'])
ser=pd.Series(data,index=[1,2,3,4,5,6,7,8])
print(ser)
print(ser[5])
print(ser[6])
```

```python
# accessing a multiple element using index lable 
data=np.array(['naveena','mohana','kani','kpr','cse','k','t','m'])
ser=pd.Series(data,index=[1,2,3,4,5,6,7,8])
print(ser[[6,7,4]])
```
